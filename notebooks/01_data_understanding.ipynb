```python
import pandas as pd
```

```python
df = pd.read_excel("../data/raw/Oil well.xlsx", header=2)
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace("\n", " ")
df.columns = df.columns.str.strip()
df.columns
```

```python
df = df.rename(columns={
    "Date": "date",
    "Oil volume (m3/day)": "oil_volume",
    "Volume of liquid (m3/day)": "liquid_volume",
    "Gas volume (m3/day)": "gas_volume",
    "Water volume (m3/day)": "water_volume",
    "Water cut (%)": "water_cut",
    "Working hours": "working_hours",
    "Dynamic level (m)": "dynamic_level",
    "Reservoir pressure (atm)": "reservoir_pressure"
})
```

```python
df.columns
```

```python
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
df.head()
```

```python
df.to_csv("../data/processed/oil_well_clean.csv", index=False)
```
